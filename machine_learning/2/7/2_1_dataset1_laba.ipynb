10. Поиск неявных дубликатов может быть проведен так

```python
def clean(response): 
    if isinstance(response, str):
        return response[0].upper() + response[1:].lower()
    return response

features["thumbnail"] = features["thumbnail"].apply(lambda x: clean(x)) 
print(features.head())
```

7. Различные типы группировок: по столбцам, по уровням индекса

```python
# По столбцам
group = dt.groupby(['average_rating', 'published_year'])
group.size()
```

```python
# По уровням индекса
copy_dt = dt.copy()
copy_dt = copy_dt.set_index(['average_rating', 'published_year'])
group = copy_dt.groupby(level=['average_rating'])
group.size()
```

8. Группировка по одному или нескольким уровням индекса

```python
# Один уровень индекса
copy_dt = dt.copy()
copy_dt = copy_dt.set_index(['average_rating', 'published_year'])
group = copy_dt.groupby(level=['average_rating'])
group.size()
```

```python
# Несколько уровней индекса
copy_dt = dt.copy()
copy_dt = copy_dt.set_index(['average_rating', 'published_year'])
group = copy_dt.groupby(level=['average_rating', 'published_year'])
group.size()
```

10. Преобразование групп: метод transform, метод apply

```python
dt.groupby('published_year').transform(sum)
```

```python
def more_than_2000(published_year):
    if published_year >= str(2000.0):
        return 'Да'
    else:
        return 'Нет'
dt['title'].apply(more_than_2000)
```
